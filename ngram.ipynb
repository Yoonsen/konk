```python
import networkx as nx

def shortest(g, source = None, target = None):
    try:
        r = nx.shortest_path(g, source, target)
    except:
        r = []
    return r

def simple(g, source = None, target = None, cutoff = 1):
    try:
        r = nx.all_simple_paths(g, source, target, cutoff = cutoff)
    except:
        r = []
    return r

def paths(g, source, target):
    short = shortest(g, source = source, target = target)
    short_opposite = shortest(g, source = target, target = source)
    short_undirected = shortest(g.to_undirected(), source = source, target = target)
    return {'shortest':{'from_source':short, 'from_target': short_opposite},
           'undirected': short_undirected}

def simplexes(g, source, target, cutoff = 4, breakpoint = 20):
    i = 0
    res = []
    for x in nx.all_simple_paths(g.to_undirected(), source= source, target = target, cutoff = 4):
        res.append(x)
        if i > breakpoint:
            break
    return res
```
